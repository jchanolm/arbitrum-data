```python
import requests
import json 
import os
import pandas as pd
from neo4j import GraphDatabase
from dotenv import load_dotenv

load_dotenv()

TALLY_API_KEY = os.getenv('TALLY_API_KEY')
```

```python
headers = {
    "Content-Type": "application/json",
    "Api-Key": TALLY_API_KEY
}

query = """
query Organization($input: OrganizationInput!) {
  organization(input: $input) {
    id
    slug
    name
    metadata {
      color
      contact {
        email
        discord
      }
      description
      icon
      socials {
        twitter
        github
      }
    }
    creator {
      id
      address
      ens
      twitter
      name
      bio
      participations {
        count
      }
      picture
      activity {
        __typename
        ... on Proposal {
          id
          title
        }
        ... on Vote {
          id
          support
        }
      }
      safes
      type
      votes
      proposalsCreatedCount
    }
    proposalsCount
    activeProposalsCount
    votersCount
    tokenHoldersCount
    chainIds
    governorIds
  }
}
"""

# Set the variables for your query. Adjust according to the organization you're querying.
variables = {
    "input": {
        "slug": "arbitrum"  # or use "id": "your_organization_id" if you have it
    }
}

# The endpoint URL for the Tally API
url = "https://api.tally.xyz/graphql"

# Make the POST request
response = requests.post(url, json={'query': query, 'variables': variables}, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()
    # Pretty print the data
    print(json.dumps(data, indent=2))
else:
    print(f"Query failed with status code {response.status_code}: {response.text}")
```
